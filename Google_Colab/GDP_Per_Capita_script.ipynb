```python
import pandas as pd


file_path = 'GDP/World_Bank_GDP.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
df.head()
```

```python
# Country names
countries = ['Mexico', 'El Salvador', 'Guatemala', 'Honduras']

# Iterating through each row and create a new DataFrame for each country
for i, country in enumerate(countries):
    # Extract row data
    row_data = df.iloc[i:i+1]
    country_df = pd.DataFrame(row_data)
    globals()[country.replace(' ', '_')] = country_df

# print(Mexico)
# print(El_Salvador)
# print(Guatemala)
# print(Honduras)
Mexico
```
